```python
import pandas as pd
import numpy as np
import os
```

```python
lsm = "MH-MH-REM-16-MS-1"
base_dir = "E:\AD_TEMP"  # "Output"
dir_list = []
for datetime in os.listdir(base_dir):
    hour_dir = os.path.join(base_dir, datetime)
    lsm_dir = os.path.join(hour_dir, lsm)
    if os.path.isdir(lsm_dir):
        try:
            dir_list += [lsm_dir]  # "Detections_" + lsm + ".xlsx"
        except:
            dir_list = [lsm_dir]
dir_list
```

```python
SHEET_NAME_ANOMALY = 'Anomaly_Report'
list_df_hr = []
for dirname in dir_list:  # Check if threading can be useful.
    detection_file = os.path.join(dirname, os.listdir(dirname)[0])
    list_df_hr += [pd.read_excel(detection_file, sheet_name=SHEET_NAME_ANOMALY)]
list_df_hr
```

```python
df_hr = pd.concat(list_df_hr)
```

```python
col_list = []
for i, col in enumerate(df_hr.columns):
    col_list += ["col" + str(i)]
df_hr.columns = list(df_hr.columns[:3]) + col_list[:-3]
df_hr
```

```python
df_hr.set_index(keys=['Short name', 'Hours'], inplace=True)
df_hr = df_hr.drop(columns=['Date'], axis=1)
```

```python
df_mean = df_hr.groupby(['Short name', 'Hours']).mean()
df_mean
```

```python
df_sum = df_hr.groupby(['Short name', 'Hours']).sum()
df_sum
```

```python
df_powsum = df_hr.groupby(['Short name', 'Hours']).apply(lambda x: np.square(x).sum())
df_powsum
```

```python
df_count = pd.DataFrame(df_hr.groupby(['Short name', 'Hours']).apply(lambda x: len(x)))
df_count.columns = ['count']
df_count
```

```python
df_org_cumsum = df_sum
df_org_powsum = df_powsum
df_org_count = df_count
```

```python
df_org_mean = pd.DataFrame(df_org_cumsum.values / df_org_count.values, index=df_org_cumsum.index,
                           columns=df_org_cumsum.columns)
df_org_mean
```

```python
df_org_var = (df_org_powsum - (df_org_cumsum * df_org_mean)) / (df_org_count.values - 1)
# df_org_var = E(X2) - E(X)^2 = S(N2)/N - M^2
# (S(N^2) + N.M^2 - 2.M.S(N)) / (N-1) where M(NM - 2S)/N = -SM/N
df_org_var
```

```python
df_org_sd = np.sqrt(df_org_var)
df_org_sd
```
